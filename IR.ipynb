```python
import codecs
import os
import sys
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def get_document_filenames(document_path='/home/isham/Projects/TFIDF'):
    return [os.path.join(document_path, each)
        for each in os.listdir(document_path)]
```

```python
Vector = TfidfVectorizer(input='filename',decode_error='ignore',stop_words="english")
tfidf_result = Vector.fit_transform(get_document_filenames())
```

```python
def display_scores(vectorizer, tfidf_result):
    # http://stackoverflow.com/questions/16078015/
    scores = zip(vectorizer.get_feature_names(),
                 np.asarray(tfidf_result.sum(axis=0)).ravel())
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    with open('su.txt','w') as f:
        for item in sorted_scores:
            print ("{0:50} Score: {1}".format(item[0], item[1]))
            f.write(str("{0:50} Score: {1}".format(item[0], item[1])) + '\n')
        
```

```python
display_scores(Vector,tfidf_result)
```
